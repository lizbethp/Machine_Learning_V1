```python
import pandas as pd
music_data = pd.read_csv('music1.csv')
X = music_data.drop(columns=['genre'])
Y = music_data['genre']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier ()
model.fit(X,Y)
predictions = model.predict([[21,1], [22,0]])
```
